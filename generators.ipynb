# Generators

A generator is a function that for a sequence of identical calls produces a sequence of different results. http://anandology.com/python-practice-book/iterators.html

```python
def yrange(n):
    i = 0
    while i < n:
        yield i
        i += 1

for i in yrange(10):
    print(i, end=' ')
```
